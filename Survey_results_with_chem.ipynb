## Chemical Signiture densities
Visualization of category acerage colored by description. a selection on category that changes the measurements on a barplot below.

```python
import pandas as pd
import numpy as np
import altair as alt
import ipywidgets as widgets
from IPython.display import display
import warnings
warnings.filterwarnings('ignore')

# enable correct rendering
alt.renderers.enable('default')

# uses intermediate json files to speed things up
alt.data_transformers.enable('json')


pd.set_option('display.max_columns', None)
```

```python
chem_data = pd.read_csv('assets/chem_data_merged.csv', index_col=0)
survey_data = pd.read_csv('assets/combined_tables.csv', index_col=0)
print('The size of the chem_data dataframe')
print('columns:', chem_data.shape[1])
print('rows   :', chem_data.shape[0])
display(chem_data.sample(5))
print('The size of the survey_data dataframe')
print('columns:', survey_data.shape[1])
print('rows   :', survey_data.shape[0])
survey_data.sample(5)
```

```python
# first we need to melt the survey data
chem_data['VisitDate'] = pd.to_datetime(chem_data['VisitDate'])
chem_data_to_merge = chem_data[((chem_data['VisitDate'].dt.year <= 2016) & (chem_data['VisitDate'].dt.year >= 2013))]

survey_data_melted = pd.melt(survey_data.drop(['from_file', '2016_population', 'Shape_Area', 'Shape_Length'], 
                            axis=1), id_vars=['LakeID', 'Description', 'Lat', 'Long', 'Town'],
                            var_name='category',
                            value_name='acres')
print(survey_data_melted.shape)
survey_data_melted.head()
```

```python
# combine with chem_data
data = survey_data_melted.merge(chem_data_to_merge, on=['LakeID', 'Lat', 'Long', 'Town'], how='inner')
data = data[data['acres'] != 0]
print(data.shape)
data.head()
```

```python
select = alt.selection_single(fields=['category'])
select_opacity = alt.condition(select, alt.value(1), alt.value(.2))

select_measure = alt.selection_single(fields=['CharacteristicID'])
select_chem_opacity = alt.condition(select_measure, alt.value(1), alt.value(.2))


survey_for_chart = data[['Description', 'category', 'acres']].drop_duplicates().groupby(['Description', 'category'])\
    .sum().reset_index()

survey_for_result = data[['LakeID', 'Description', 'category', 'CharacteristicID', 'CharacteristicName', 'UnitCode','NormResult']].drop_duplicates()\
    .groupby(['LakeID', 'Description', 'category', 'CharacteristicID', 'CharacteristicName', 'UnitCode'])\
    .mean().reset_index()


acres = alt.Chart(survey_for_chart).mark_bar().encode(
    x=alt.X('category:N', sort='x'),
    y=alt.Y('sum(acres):Q'),
    color=alt.Color('Description:N'),
    tooltip=[alt.Tooltip('category'), alt.Tooltip('acres', title='total acres')],
    opacity=select_opacity
).properties(
    width=600
).add_selection(
    select
).interactive()

measures = alt.Chart(survey_for_result).mark_bar(strokeWidth=20, opacity=.001).encode(
        x=alt.X('CharacteristicID:N', sort='x'),
        y=alt.Y('mean(NormResult):Q', scale=alt.Scale(domain=[0, 1])),
        tooltip=[alt.Tooltip('CharacteristicName'), alt.Tooltip('UnitCode'), alt.Tooltip('mean(NormResult)', title='Mean NormResult')],
        opacity=select_chem_opacity
    ).properties(
        width=600
    ).add_selection(
        select_measure
    ).transform_filter(
        select
    ).interactive()

(acres & measures)
```
